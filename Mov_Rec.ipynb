```python
#Import the required libraries

import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#Reading the dataset

data_mov = pd.read_csv("movies.csv")
```

```python
data_mov.head()
```

```python
#Selecting the required features

selected_features = ["genres","keywords","tagline","cast","director"]
```

```python
#Cleaning the Selected data

for features in selected_features:
    data_mov[features] = data_mov[features].fillna("") 
```

```python
#Combining the Selected data

comb_features = data_mov["genres"]+''+data_mov["keywords"]+''+data_mov["tagline"]+''+data_mov["cast"]+''+data_mov["director"]
```

```python
print(comb_features)
```

```python
#Creating an instance of the vectorizer

vectorizer = TfidfVectorizer()
```

```python
#Training and fitting the data

feature_vectors = vectorizer.fit_transform(comb_features)
```

```python
print(feature_vectors)
```

```python
#Generating the similarity scores

similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
#Getting the movie name from user

movie_name = input("Enter your favourite movie name:")
```

```python
#List of the movie titles

list_titles = data_mov["title"].tolist()
```

```python
print(list_titles)
```

```python
#Finding the close match for thee movie name given by user

find_close_match = difflib.get_close_matches(movie_name, list_titles)
```

```python
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
#Finding the index of the movie with title

index_movie = data_mov[data_mov.title == close_match]['index'].values[0]
print(index_movie)
```

```python
#Getting list of similar movies

simi_score = list(enumerate(similarity[index_movie]))
print(simi_score)
```

```python
len(simi_score)
```

```python
#Sorting the movies based on the similarity score

sort_simi_movies = sorted(simi_score, key = lambda x:x[1], reverse = True)
print(sort_simi_movies)
```

```python
#final

print("Movies recommended for you: \n")

i=1
for movie in sort_simi_movies:
    index = movie[0]
    title_from_index = data_mov[data_mov.index==index]["title"].values[0]
    if (i<10):
        print(i,".",title_from_index)
        i+=1
```

```python
# Movie Recommendation System

movie_name = input("Enter your favourite movie name:")
list_titles = data_mov["title"].tolist()
find_close_match = difflib.get_close_matches(movie_name, list_titles)
close_match = find_close_match[0]
index_movie = data_mov[data_mov.title == close_match]['index'].values[0]
simi_score = list(enumerate(similarity[index_movie]))
sort_simi_movies = sorted(simi_score, key = lambda x:x[1], reverse = True)

print("Movies recommended for you: \n")

i=1
for movie in sort_simi_movies:
    index = movie[0]
    title_from_index = data_mov[data_mov.index==index]["title"].values[0]
    if (i<10):
        print(i,".",title_from_index)
        i+=1
```
